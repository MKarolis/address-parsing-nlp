Lots of labels in the excel file were not correct, causing errors when puting the entries in the right format so it could be used for the model. For that we created warnings so we could know which entries were causing problems and why. In the end we had wat we called WARNINGs and overlaps, as we can se below:



When the warning is like "WARNING: could not find token "1956-02-01 00:00:00" in address "2-56, Suda-cho, Mizuho-ku, Nagoya-shi, Aichi 467-8530"" the error is obvious. What happens is that in the excel file when we put something like 2-56 it automaticly transforms it into a Date type, meaning that it will be readed as Feb-56, like in the exemple, where the street is a Date:



For that problem, we simply pu a space in front of the information we want and then it doesn' transforms it into a Date anymore. And since we do a .strip() for every entry before it gets treated, it's not a problem to have it.
